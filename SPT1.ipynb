```python
import pandas as pd
import numpy as np
import itertools
from itertools import combinations
```

```python
dataFrame = pd.read_csv('https://raw.githubusercontent.com/Praneashram/Priority1/main/Priority2.csv')
```

```python
#Arrival time based sorting
#dataFrame = dataFrame.sort_values('Arrival Time', ascending=True, ignore_index=True)
temp = dataFrame.sort_values(['Arrival Time', 'Processing Time', 'Priority'], ascending=[True, True, True],ignore_index=True)
temp
```

```python
minArrivalTime = min(list(dataFrame['Arrival Time']))
startingProcesses = dataFrame[dataFrame['Arrival Time'] == minArrivalTime]
startingProcesses = startingProcesses.sort_values('Processing Time', ascending=True, ignore_index=True)
startingProcessesList = list(startingProcesses['Process'])
#print(startingProcessesList)
for s in startingProcessesList:
  startingIndex = dataFrame[dataFrame['Process'] == s].index
  dataFrame = dataFrame.drop(startingIndex)
dataFrame = dataFrame.sort_values('Processing Time', ascending=True, ignore_index=True)
dataFrame = startingProcesses.append(dataFrame)
dataFrame = dataFrame.reset_index(drop=True)
dataFrame
#data = dataFrame[['Process'],ignore_index=True]
#data
```

```python
completed = []
waiting = []
entryTimes = []
endTimes = []
entryTime = -1
endTime = min(temp['Arrival Time'])
while(len(completed) != len(temp['Process'])):
  for idx, val in dataFrame.iterrows():
    if(val['Process'] not in completed):
      entryTime = endTime
      #print(val['Process'])
      if val['Arrival Time'] <= entryTime:
        completed.append(val['Process'])
        entryTimes.append(entryTime)
        endTime = entryTime + val['Processing Time']
        endTimes.append(endTime)
        break
#print(completed)

```

```python
#sorted1 = pd.DataFrame(dataFrame, columns = ['Process'])
sorted1 = pd.DataFrame(entryTimes, columns = ['Entry Time'])
sorted2 = pd.DataFrame(endTimes, columns = ['End Time'])
sorted_final = pd.concat([dataFrame,sorted1,sorted2], axis=1)
sorted_final['Turnaround Time'] = sorted_final['End Time'] - sorted_final['Arrival Time']
sorted_final['Waiting Time'] = sorted_final['Entry Time'] - sorted_final['Arrival Time']
sorted_final
```

```python
waiting = []
turnaround = []
sum_waiting_time = 0
sum_turnaround_time = 0
avg_waiting_time = 0
avg_turnaround_time = 0
for index, row in sorted_final.iterrows():
  wait = row['Waiting Time']
  turn = row['Turnaround Time']
  waiting.append(wait)
  turnaround.append(turn)

waiting = np.array(waiting)
turnaround = np.array(turnaround)
#print(waiting)
#print(turnaround)
#sum of waiting time for all the processes are calculated
for i in range(waiting.shape[0]):
  sum_waiting_time+=waiting[i]
#sum of turnaround time for all the processes are calculated
for i in range(turnaround.shape[0]):
  sum_turnaround_time+=turnaround[i]

avg_turnaround_time= sum_turnaround_time/len(entryTimes)
avg_waiting_time= sum_waiting_time/len(entryTimes)

print('average waiting time =', avg_waiting_time)
print('average turnaround time =', avg_turnaround_time)
```
